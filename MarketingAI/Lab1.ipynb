```python
import string

temperatures = (23, 25, 28, 30, 29, 31, 27, 26, 30, 32, 33, 29, 28, 27, 31, 30, 29, 28, 26, 25, 24, 23, 22, 21, 20, 22, 24, 26, 27, 28, 29)

max_temp = max(temperatures)
print("Максимальная температура в августе:", max_temp)

A = (
    (1, 2, 3),
    (4, 5, 6)
)

B = (
    (7, 8),
    (9, 10),
    (11, 12)
)

rows_A = len(A)
cols_A = len(A[0])
rows_B = len(B)
cols_B = len(B[0])

if cols_A != rows_B:
    print("Матрицы нельзя перемножить")
else:
    result = []
    for i in range(rows_A):
        row = []
        for j in range(cols_B):
            s = 0
            for k in range(cols_A):
                s += A[i][k] * B[k][j]
            row.append(s)
        result.append(tuple(row))
    result = tuple(result)

    print("Результат умножения матриц:")
    for row in result:
        print(row)
numbers = [1, 2, 3, 2, 1, 1, 4, 2, 2]
count_pairs = 0
freq = {}
for num in numbers:
    freq[num] = freq.get(num, 0) + 1

for val in freq.values():
    if val > 1:
        count_pairs += val * (val - 1) // 2

print("Количество пар равных элементов:", count_pairs)

list1 = [5, 3, 9, 1, 4, 7, 8, 3, 5]
list2 = [3, 7, 10, 5, 7, 8]

set1 = set(list1)
set2 = set(list2)

common = sorted(set1.intersection(set2))

print("Общие числа в порядке возрастания:", common)

text = input("Введите строку: ")


translator = str.maketrans('', '', string.punctuation)
clean_text = text.translate(translator).lower()

words = clean_text.split()
freq_words = {}
for w in words:
    freq_words[w] = freq_words.get(w, 0) + 1

min_freq = min(freq_words.values())

rare_words = [w for w, count in freq_words.items() if count == min_freq]
answer = min(rare_words)

print("Слово, которое встречается реже всего:", answer)
```
